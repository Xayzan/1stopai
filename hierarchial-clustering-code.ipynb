Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv("Mall_Customers-211105-191711.csv")
X = dataset.iloc[:, :].values
```

```python
X
```

Dendrogram representing the optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title("Dendrogram")
plt.xlabel("Customers")
plt.ylabel("Euclidean Distance")
plt.show()
```

Training the data for the model

```python
from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering(n_clusters = 5).fit(X)
y_hc = clustering.fit_predict(X)
```

```python
y_hc 
```

Creating a visual representation of the work done on the data ie clusters

```python
from cProfile import label


plt.scatter(X[y_hc == 0, 0], X[y_hc == 0,1], c = 'red', label = 'Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1,1], c = 'orange', label = 'Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2,1], c = 'pink', label = 'Cluster 3')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3,1], c = 'green', label = 'Cluster 4')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4,1], c = 'blue', label = 'Cluster 5')
plt.title("Cluster of Customers")
plt.xlabel("Annual Income(K$)")
plt.ylabel("Spending Score(1-100)")
plt.legend()
plt.show()
```
